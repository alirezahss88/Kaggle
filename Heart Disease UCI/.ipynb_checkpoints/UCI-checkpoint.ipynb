```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from funcs import outliers
from imblearn.over_sampling import RandomOverSampler
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, recall_score, precision_score, precision_recall_curve, plot_confusion_matrix
from sklearn.ensemble import AdaBoostClassifier

sns.set()
```

# Load the dataset from the csv file

```python
data = pd.read_csv('D:\Books\Machine Learning\Real_World_project\Kaggle\Heart Disease UCI\heart.csv')
print('Dataset shape: ', data.shape)
data.head()
```

## What are the features?

***age***: The person's age in years\
***sex***: The person's sex (1 = male, 0 = female)\
***cp***: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)\
***trestbps***: The person's resting blood pressure (mm Hg on admission to the hospital)\
***chol***: The person's cholesterol measurement in mg/dl\
***fbs***: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)\
***restecg***: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)\
***thalach***: The person's maximum heart rate achieved\
***exang***: Exercise induced angina (1 = yes; 0 = no)\
***oldpeak***: ST depression induced by exercise relative to rest\
***slope***: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)\
***ca***: The number of major vessels (0-3)\
***thal***: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)\

***target***: Heart disease (0 = no, 1 = yes)

# Cleaning data
## Plot Box and Whisker
We plot whisker and box plot to identify outliers. As shown below, four features including *trestbps*, *chol*, *thalach*, and *oldpeak* have outliers.

```python
plt.figure(figsize=(20,15))
for i, f in enumerate(['age', 'trestbps', 'chol', 'thalach', 'oldpeak']):
    plt.subplot(4, 4, i+1)
    plt.boxplot(data[f], labels=[f])
```

## Removing outliers
We used Interquartile Range (IQR) technique to identify and delete outliers.

```python
for column in ['trestbps', 'chol', 'thalach', 'oldpeak']:
    X_train, y_train = outliers(X_train, y_train, column)
    
print('Dataset shape after removing outliers: ', X_train.shape)
```

Plotting Whisker and box plot to make sure that all outliers are deleted.

```python
plt.figure(figsize=(20,15))
for i, f in enumerate(['age', 'trestbps', 'chol', 'thalach', 'oldpeak']):
    plt.subplot(4, 4, i+1)
    plt.boxplot(X_train[f], labels=[f])
```

# Implementation
## Creating a StandardScaler object

```python
ssc = StandardScaler()
X_train_scaled = ssc.fit_transform(X_train)
```

## Running a Grid Search CV on a Random Forest estimator

```python
rf = RandomForestClassifier(class_weight='balanced')
param_grid = {'n_estimators': np.logspace(1, 5, 5, base=3, dtype=np.int16),
              'max_depth': np.logspace(1, 5, 5, base=3, dtype=np.int16),
              'min_samples_split': [2, 4, 6, 8, 10, 12, 14]}

gs_rf = GridSearchCV(rf, param_grid=param_grid, cv=5)
gs_rf.fit(X_train_scaled, y_train)
```

## Printing the training results

```python
print('RESULTS ON THE TRAINING SET:')
print('   the best params are: ', gs_rf.best_params_)
print('   the best accuracy is: {:.2f}%'.format(100*gs_rf.best_score_))
print('   the best precision is {:.2f}%: ' .format(100*precision_score(y_train, gs_rf.predict(X_train_scaled))))
print('   the best recall is {:.2f}%: '.format(100*recall_score(y_train, gs_rf.predict(X_train_scaled))))
print('   the best f1_score is {:.2f}%: ' .format(100*f1_score(y_train, gs_rf.predict(X_train_scaled))))
```

```python
plot_confusion_matrix(gs_rf, X_train_scaled, y_train)
plt.axis('off');
```

## Predicting and printing the scores associated with the test set

```python
X_test_scaled = ssc.transform(X_test)  #Scaling of the test set should be done by the object fitted by the training set
```

```python
print('RESULTS ON THE TEST SET:')
print('   the accuracy is: {:.2f}%'.format(100*gs_rf.score(X_test_scaled, y_test)))
print('   the precision is {:.2f}%: ' .format(100*precision_score(y_test, gs_rf.predict(X_test_scaled))))
print('   the recall is {:.2f}%: '.format(100*recall_score(y_test, gs_rf.predict(X_test_scaled))))
print('   the f1_score is {:.2f}%: ' .format(100*f1_score(y_test, gs_rf.predict(X_test_scaled))))
```

```python
ax = plot_confusion_matrix(gs_rf, X_test_scaled, y_test, display_labels=True);
plt.axis('off');
```
